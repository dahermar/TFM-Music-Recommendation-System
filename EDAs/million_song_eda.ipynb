```python
df_music.describe()
```

```python
correlation_matrix = df_music.select_dtypes(include='number').corr()
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show()
```

Observable correlations:
- Positive correlation of 0.78 between loadness and energy and 0.54 between danceability and valence.
- Negative correlation of -0.73 betweeen acousticness and energy and -0.61 between loudness and acousticness.

```python
# Graficar distribuciones de variables numÃ©ricas
numeric_columns = df_music.select_dtypes(include=['number']).columns.tolist()

plt.figure(figsize=(20, 15))
for i, column in enumerate(numeric_columns, 1):
    plt.subplot(4, 4, i)
    sns.histplot(df_music[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
numeric_columns = df_music.select_dtypes(include=['number']).columns.tolist()
for col in ['energy', 'year']:
    if col in numeric_columns:
        numeric_columns.remove(col)

num_plots_per_figure = 4
num_figures = 3

for fig_num in range(num_figures):
    plt.figure(figsize=(15, 10))
    start_idx = fig_num * num_plots_per_figure
    end_idx = start_idx + num_plots_per_figure
    for i, column in enumerate(numeric_columns[start_idx:end_idx], 1):
        plt.subplot(2, 2, i)
        sns.histplot(df_music[column], kde=True)
        plt.title(f'Distribution of {column}', fontsize=22)
        plt.xlabel(column, fontsize=16)
        plt.ylabel('Frequency', fontsize=16)
        plt.xticks(fontsize=14)
        plt.yticks(fontsize=14)
    plt.tight_layout()
    plt.savefig(f'distributions_part_{fig_num + 1}.png')
    plt.close()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df_music['energy'], kde=True)
plt.title('Distribution of energy in Million Song Dataset + Spotify + Last.fm')
plt.xlabel('energy')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
default_colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
genre_counts = df_music['genre'].value_counts()
palette = default_colors[:len(genre_counts)]

plt.figure(figsize=(10, 6))
sns.barplot(x=genre_counts.index, y=genre_counts.values, hue=genre_counts.index, palette='husl')

plt.xticks(rotation=45)
plt.xlabel("Genre")
plt.ylabel("Number of songs")
plt.title("Number of songs per genre")

plt.show()
```

## Users

```python
df_users = pd.read_csv('../System/data/Million Song Dataset kaggle/User Listening History.csv')
```

```python
df_users
```

```python
num_plays = df_users['playcount'].sum()
print(f'Total number of plays: {num_plays}')
```

```python
df_users.dtypes
```

```python
num_distinct_users = df_users['user_id'].nunique()
num_distinct_users
```

```python
num_distinct_tracks = df_users['track_id'].nunique()
num_distinct_tracks
```

```python
data_df = df_users.groupby('track_id').size().reset_index(name='count')
```

```python
data_df
```

Most played songs by all users

```python
df_music_count = df_music_info.merge(
    df_users.groupby('track_id').size().reset_index(name='playcount'),
    on='track_id',
    how='left'
)
df_music_count = df_music_count.fillna(0)
print(df_music_count.shape)
df_music_count.head()
```

```python
played_song_mean = df_users.groupby('user_id')['track_id'].nunique().mean()
print(f"Mean of tracks played by user: {played_song_mean:.2f}")
```

```python
df_user_counts = df_users['user_id'].value_counts().reset_index()
df_user_counts.columns = ['user_id', 'count']

num_users_less_5_songs = df_user_counts[df_user_counts['count'] <= 5].reset_index(drop=True).shape[0]

print(f"Percentage of users with less than 5 songs: {num_users_less_5_songs / num_distinct_users:.2%}")
```

```python
print(f"Number of distinct users: {num_distinct_users}")
print(f"Number of users with more than 5 songs: {num_distinct_users - num_users_less_5_songs}")
```

```python
playcount_users_less_5 = df_users[df_users['user_id'].isin(df_user_counts[df_user_counts['count'] <= 5]['user_id'])].reset_index(drop=True)
playcount_users_less_5['playcount'].sum()
print(f"Percentage of playcount from users with less than 5 songs: {playcount_users_less_5['playcount'].sum() / num_plays:.2%}")
```

```python
df_users_agg = df_users.groupby('user_id')['playcount'].agg(
    #total_playcount='sum',
    max_playcount='max'
).reset_index()
df_users_agg = df_users_agg.rename(columns={'playcount': 'max_playcount'})
```

```python
df_users_rating = df_users.merge(df_users_agg, on='user_id')
```
